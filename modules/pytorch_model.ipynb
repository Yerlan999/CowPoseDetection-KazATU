```python
import torch as t
import torchvision as tv
import xml.etree.ElementTree as ET
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
xml_data = open(r'cows/train/annotations/co1.xml').read()
root = ET.XML(xml_data)  

data = []
columns=["keypoint", 'is_visible', "x", "y", "z"]

for keypoint in root[3]:
    name, visible, x, y, z = keypoint.items()
    keypoint_name = name[1]; is_visible = visible[1]; x = x[1]; y = y[1]; z = z[1];
    data.append([keypoint_name, is_visible, x, y, z])
    
df = pd.DataFrame(data, columns=columns)
df
```
